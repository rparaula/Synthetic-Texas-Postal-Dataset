# Mapping multiple Olist sources codes to a Texas ZIP based on a cumulaive capacity packing algorithm

The algorithm is basically:

1. Give every source_code a size based on Olist customer count.
2. Give every Texas ZIP a capacity based on population/density weighting.
3. Sort both sides from largest to smallest.
4. Walk through the cumulative distributions.
5. Assign each source_code to the Texas ZIP whose capacity range contains it.
6. Join the mapping back to the full customer table.
