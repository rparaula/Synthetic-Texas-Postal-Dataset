# Build P2P Olist Origin Ingestion

This notebook builds the first-pass P2P Olist origin ingestion artifacts for the beginning package lifecycle only.


## What This Produces

- `stg_p2p_package.csv`
- `stg_p2p_shippingdetails.csv`
- `stg_p2p_origin_movement.csv`
- `stg_p2p_package_quarantine.csv`

Package weight is staged in ounces to match the production `package.weight_oz` column.


```python
from p2p_olist_origin_pipeline import (
    OUTPUT_DIR,
    QUARANTINE_CSV,
    STG_MOVEMENT_CSV,
    STG_PACKAGE_CSV,
    STG_SHIPPINGDETAILS_CSV,
    TARGET_P2P_FINAL_SHARE,
    TARGET_P2P_PACKAGE_COUNT,
    build_all_outputs,
    print_validation_report,
)
```

```python
TARGET_P2P_FINAL_SHARE = 0.25
TARGET_P2P_PACKAGE_COUNT = None
```

```python
result = build_all_outputs(
    write_files=True,
    target_p2p_final_share=TARGET_P2P_FINAL_SHARE,
    target_p2p_package_count=TARGET_P2P_PACKAGE_COUNT,
)
print_validation_report(result)
```

```python
print(f'Output directory: {OUTPUT_DIR}')
print(f'Package staging CSV: {STG_PACKAGE_CSV.exists()} -> {STG_PACKAGE_CSV.name}')
print(f'Shippingdetails staging CSV: {STG_SHIPPINGDETAILS_CSV.exists()} -> {STG_SHIPPINGDETAILS_CSV.name}')
print(f'Movement staging CSV: {STG_MOVEMENT_CSV.exists()} -> {STG_MOVEMENT_CSV.name}')
print(f'Quarantine CSV: {QUARANTINE_CSV.exists()} -> {QUARANTINE_CSV.name}')
```
